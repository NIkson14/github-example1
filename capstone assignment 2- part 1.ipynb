# IMPORTING ALL REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import requests
```

```python
from bs4 import BeautifulSoup
```

## 1) Scraping Content From Wiki Page

```python
# getting data
# get html from wiki page and create soup object
source = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(source.text, 'lxml')
```

```python
# scraping table & transforming data into datafrme
table = soup.find('table',{'class':'wikitable sortable'})
table_rows = table.find_all('tr')
data= []
for row in table_rows:
    data.append([t.text.strip() for t in row.find_all('td')])
```

```python
df = pd.DataFrame(data, columns=['PostalCode', 'Borough', 'Neighbourhood'])
df = df[~df['PostalCode'].isnull()]
df.head(10)
```

```python
# data cleanup
df.drop(df[df['Borough']=="Not assigned"].index,axis=0, inplace=True)
df.head(10)
```

```python
# grouping all neighborhoods with same postal code

df = df.groupby('PostalCode').agg(lambda x: ','.join(x))

# for cells that have an assigned neighbouhoods,like M7A,assigning their boroughs as their neighbourhood, 

df.loc[df['Neighbourhood']=="Not assigned",'Neighbourhood']=df.loc[df['Neighbourhood']=="Not assigned",'Borough']
df=df.reset_index()
df['Borough']= df['Borough'].str.replace('nan|[{}\s]','').str.split(',').apply(set).str.join(',').str.strip(',').str.replace(",{2,}",",")
df.head(10)
```
